# Visualisation

```python
status=df.groupby(["status"]).size().reset_index().rename(columns={0:"Count"})
```

```python
status_names=status.status.value_counts().index
status_names
```

```python
status_val=status.Count.value_counts().index
status_val
```

```python
plt.pie(status_val, labels=status_names, autopct="%1.2f%%")
plt.title('Distribution of the Status of Candidates')
plt.plot()
```

# Relationship between Degree Specialisation and Candidate placement

```python
df["specialisation"].unique()
```

```python
sns.countplot(x="status", data=df, hue='specialisation')
plt.title("Degree Specialization vs Candidate Placement")
plt.xlabel("Status of Placement-1 is Placed and 0 is Not Placed")
plt.ylabel("Number of candidate")
plt.show()
```

# Observation 2

# Relationship between Work Exp and Candidate placement¶

```python
sns.countplot(x="status", data=df, hue='workex')
plt.title("Work Exp vs Candidate Placement")
plt.xlabel("Status of Placement-1 is Placed and 0 is Not Placed")
plt.ylabel("Number of candidate")
plt.show()
```
